```python
# Sample DataFrame structure
df = pd.DataFrame({
        'Quantity': [10, 20, 30, 40, 50],
        'TotalSales': [100, 200, 300, 400, 500]
    })
```

```python
plt.figure(figsize=(10, 6))  # Set the size of the figure

# Create the scatter plot
sns.scatterplot(data=df, x='Quantity', y='TotalSales')

# Set plot title and labels
plt.title('Quantity vs Total Sales', fontsize=16)
plt.xlabel('Quantity', fontsize=12)
plt.ylabel('Total Sales (in USD)', fontsize=12)

# Optional: Add a regression line
sns.regplot(data=df, x='Quantity', y='TotalSales', scatter=False, color='r')

# Adjust layout
plt.tight_layout()

# Save the plot as an image
plt.savefig('quantity_vs_sales_scatter.jpg', format='jpg')

# Display the plot
plt.show()
```
